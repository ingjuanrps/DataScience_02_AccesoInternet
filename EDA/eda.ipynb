```python
# Filtramos outliers de totalServicios, para apreciar mejor:
# El 25% de nuestros datos inferiorioes a la media. -4.0
# La media que es el 50% de nuestros datos. El rango intercuartil (IQR) de total_aboard es 4.0 entre -4.0  y 12.0
# El otro 25% de nuestors datos que estan por encima de la media. 12.0
sns.boxplot(data = internet, x = 'totalServicios')
plt.savefig(r'C:\Users\ingju\Desktop\JrPs_PI2\src\outliers2', dpi=50)
plt.show()
# IQR
Q1 = np.percentile(internet['totalServicios'], 25,
                   interpolation = 'midpoint')
Q3 = np.percentile(internet['totalServicios'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
print(f"El rango intercuartil (IQR) de total_aboard es {IQR}")

lower_lim = Q1 - 1.5*IQR
upper_lim = Q3 + 1.5*IQR
print(lower_lim, upper_lim)

internet[(internet['totalServicios']<lower_lim) | (internet['totalServicios']>upper_lim)]
```

```python
# Criterio de selección.

internet[(internet['totalServicios']>12)]
```

## Metricas.

¿Cuantas Provincias estamos análizando?

```python
provincias=internet['Provincia'].nunique()
print('El total de provincias análizadas son: ', provincias)
```

¿Que provincia tiene mayor acceso al internet?

```python
resultado=internet.groupby('Provincia')['AccesoCada100Hogares'].agg(['sum','mean']).reset_index()
resultado_o=resultado.sort_values(by='sum',ascending=False)
print('La Provincia con mayor acceso a internet es: ')
print(resultado_o)
```

```python
# Crear una figura y ejes para el gráfico
fig, ax = plt.subplots(figsize=(10, 6))

# Barra para la suma de calificaciones
ax.bar(resultado_o['Provincia'], resultado_o['sum'], label='Suma de Calificaciones', color='blue')

# Línea para el promedio de calificaciones
ax.plot(resultado_o['Provincia'], resultado_o['mean'], label='Promedio de Calificaciones', color='red', marker='o')

# Personalizar el gráfico
ax.set_xlabel('Provincias')
ax.set_ylabel('Acceso a Internet')
ax.set_title('Acceso a Internet cada 100 Hogares por Provincia')
ax.legend()

# Rotar los nombres en el eje x para una mejor visualización
plt.xticks(rotation=90)

# Mostrar el gráfico
plt.tight_layout()
plt.show()

plt.savefig(r'C:\Users\ingju\Desktop\JrPs_PI2\src\barras2', dpi=50)
```

¿Qué año tuvo mayor crecimiento con forme a los 100 hogares por trimestre?

```python
# Calcular la calificación promedio por año y trimestre
calificacion_promedio_anio = internet.groupby("Anio")["AccesoCada100Hogares"].mean()
calificacion_promedio_trimestre = internet.groupby(["Anio", "Trimestre"])["AccesoCada100Hogares"].mean()

# Encontrar el mejor año y trimestre
mejor_anio = calificacion_promedio_anio.idxmax()
mejor_trimestre = calificacion_promedio_trimestre.idxmax()

# Crear un gráfico de pastel para mostrar la distribución de calificaciones por año
plt.figure(figsize=(10, 12))
plt.pie(calificacion_promedio_anio, labels=calificacion_promedio_anio.index, autopct='%1.1f%%')
plt.title("Distribución de Acceso a Internet por Año")
plt.show()

print("El mejor año en términos de calificación promedio fue:", mejor_anio)
```
